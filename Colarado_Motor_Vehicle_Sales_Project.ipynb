```python
plt.figure(figsize=(10,7))
sns.lineplot(data=motor_df,x='period',y='sales')
plt.xticks(rotation=90)
plt.show()
```

```python
# sales analysis yearly and quarterly on the basis of country
plt.figure(figsize=(10,7))
sns.lineplot(data=motor_df,x='period',y='sales',hue='county')
plt.xticks(rotation=90)
plt.show()
```

```python
# box plot
plt.figure(figsize=(10,7))
sns.boxplot(data=motor_df,x='quarter',y='sales')
plt.title('Sales Distribution by Quarter')
plt.show()
```

```python
year=int(input('Enter The Year Here:'))
Quarter=int(input('Enter The Quarter Here:'))
def plot_sales_by_county(df,year,quarter):
  filtered_df=df[(df['year']==year)&(df['quarter']==quarter)]
  filtered_df.groupby(by='county')['sales'].sum().sort_values(ascending=False).plot(kind='bar',width=0.8)
plot_sales_by_county(motor_df,year,Quarter)
```
